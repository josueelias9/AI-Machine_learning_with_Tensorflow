```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
print(tf.__version__)
```

```python
aa = tf.keras.datasets.fashion_mnist

(train_data, train_label), (test_data,test_label)=aa.load_data()
```

```python
train_data = train_data / 255.0
test_data = test_data / 255.0
```

```python

# plt.colorbar()
plt.figure(figsize=(10,10))
for i in range(16):
  plt.subplot(4,4,i+1)
  plt.imshow(train_data[i])
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(128, activation='relu',use_bias=True),
    tf.keras.layers.Dense(10, activation='relu')
]
)
model.summary()

model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)
model.fit(
    train_data, 
    train_label,
    epochs=10)
```

```python
oeoe = model.predict(test_data)
print(oeoe[0])
print(test_label[0])

print(oeoe[1])
print(test_label[1])
```

```python
model_2 = tf.keras.Sequential([
    model,
    tf.keras.layers.Softmax()
])
oeoe2 = model_2.predict(test_data)
oeoe2[0]
model_2.summary()
```
